```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

def ReLU(x):
    result = []
    for i in x:
        if i < 0:
            result.append(0)
        else:
            result.append(i)

    return result

x = np.linspace(-10, 10)
plt.plot(x, ReLU(x))
plt.axvline(0.0, color = 'k',linestyle = '--')
plt.xlabel('z')
plt.ylabel('$activation (z) $')
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.title('Activation Function :ReLU')
plt.show()
```
