```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
x = (-5,-1,0,0,1,5)
y = (0,0,0,1,1,1)
plt.plot(x, y)
plt.axvline(0.0, color = 'k',linestyle = '--')
plt.xlabel('z')
plt.ylabel('$activation (z) $')
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.title(r"Activation Function :Threshold $\Theta = 1$")
plt.show()
```
