```python
import matplotlib.pyplot as plt 
import numpy as np
z = np.arange(-5,5,0.1)
phi_z = 1.0/(1.0 + np.exp(-z))
plt.plot(z, phi_z)
plt.axvline(0.0, color = 'k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\sigma (z) $')
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```
